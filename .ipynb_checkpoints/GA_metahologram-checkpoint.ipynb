```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors,cm
import math
import cv2
import torch.nn.functional as F
import torch
```

```python
img = cv2.imread('.\\five.png')
groundtruth = cv2.cvtColor(img,cv2.COLOR_RGBA2GRAY)
groundtruth = np.double(groundtruth/255)
plt.imshow(groundtruth)
#print(np.max(farfield))
```

```python
# 分成八阶相位
StepPhase = np.array([0,0.125,0.25,0.375,0.5,0.625,0.75,0.875])
StepPhaseMid = np.zeros([7,])
for i in range(7):
    StepPhaseMid[i] = (StepPhase[i] + StepPhase[i+1]) / 2


Phase = groundtruth
initatialPhase = np.copy(Phase)
for l in range(initatialPhase.shape[0]):
    for m in range(initatialPhase.shape[1]):
        for n in range(7):
            if(0 <= groundtruth[l][m] < StepPhaseMid[n]):
                initatialPhase[l][m] = StepPhase[0]
            elif(groundtruth[l][m] > StepPhaseMid[6]):
                initatialPhase[l][m] = StepPhase[7]
            elif(StepPhaseMid[n-1] <= groundtruth[l][m] < StepPhaseMid[n]):
                initatialPhase[l][m] = StepPhase[n]

plt.imshow(initatialPhase)
print(np.max(initatialPhase*8))
print(StepPhaseMid)
```

```python
# metasurface参数
N = 100
STEP = 3
# GA参数
DNA_SIZE = N*N*STEP  #100*100的超表面，每个分成八阶相位
CROSS_RATE = 0.8
MUTATE_RATE = 0.1
POP_SIZE = 10
N_GENERATIONS = 100
# 光学参数
lambdad = 532e-9
d = 40e-6
L = 100 * 0.4e-9
```

```python
# 遗传算法，变量太多，收敛极慢
class MGA(object):
    def __init__(self, DNA_size, DNA_bound, cross_rate,mutation_rate,pop_size,lambdad,d,L):
        self.DNA_size = DNA_size
        DNA_bound[1] += 1
        self.DNA_bound = DNA_bound
        self.cross_rate = cross_rate
        self.mutate_rate = mutation_rate
        self.pop_size = pop_size
        self.lambdad = lambdad
        self.dist = d
        self.L = L

        # initial DNAs for winner and loser   0 1   100 100 3
        self.pop = np.random.randint(*DNA_bound,size=(1,N,N,STEP)).repeat(self.pop_size, axis=0)
        
    def AS_Propagation(self,Uin):

        nx = Uin.shape[0]
        ny = Uin.shape[1]
        x = np.linspace(-1/2*self.L,1/2*self.L,nx)
        y = np.linspace(-1/2*self.L,1/2*self.L,ny)
        X,Y=np.meshgrid(x,y)
        fx = X / self.L
        fy = Y / self.L
        q = fx**2 + fy**2
        k = 2*np.pi/self.lambdad
        r = np.sqrt(X**2+Y**2+self.dist**2)
        H = np.exp(1j*k*self.dist*np.sqrt(1-self.lambdad**2*q))
        O = np.fft.fftshift(np.fft.fft2(np.fft.fftshift(Uin)))
        U = np.fft.fftshift(np.fft.ifft2(np.fft.fftshift(H*O)))

        return U

    def MSE(self,pred,target):
        return (np.sum((target - pred) ** 2)/len(target))
    
    def translateDNA(self,pop):
        c = pop.dot(2 ** np.arange(STEP)[::-1])/2**STEP * 2*np.pi
        return c

    def get_fitness(self, product):
        field = np.empty(product).astype(np.complex128)
        I = np.empty(product).astype(np.float64)
        fitness = np.empty(product).astype(np.float64)
        for i in range(product.shape[2]):
            phase = product[:,:,i]
            field[:,:,i] = self.AS_Propagation(1*np.exp(1j*phase))
            I[:,:,i] = np.real(np.sqrt(field[:,:,i]*np.conj(field[:,:,i])))
            I[:,:,i] = I[:,:,i]/np.max(I[:,:,i])
            fitness[i] = self.MSE(I[:,:,i],groundtruth)
        return fitness

    def crossover(self, loser_winner):      # crossover for loser
        cross_idx = np.empty((self.DNA_size,)).astype(bool)
        for i in range(self.DNA_size):
            cross_idx[i] = True if np.random.rand() < self.cross_rate else False  # crossover index
        loser_winner[1, cross_idx] = loser_winner[0, cross_idx]  # assign winners genes to loser
        return loser_winner

    def mutate(self, loser_winner):         # mutation for loser
        mutation_idx = np.empty((self.DNA_size,)).astype(bool)
        for i in range(self.DNA_size):
            mutation_idx[i] = True if np.random.rand() < self.mutate_rate else False  # mutation index
        loser_winner[1, mutation_idx] = ~loser_winner[1, mutation_idx].astype(bool)
        return loser_winner

    def evolve(self, n):
        for _ in range(n):
            sub_pop_idx = np.random.choice(np.arange(0,self.pop_size), size=2, replace=False) #随机选两个
            sub_pop = self.pop[sub_pop_idx,:,:,:]
            product = self.translateDNA(sub_pop) #10 100 100 
            #print(sub_pop.shape)
            fitness = self.get_fitness(product)
            loser_winner_idx = np.argsort(fitness)
            loser_winner = sub_pop[loser_winner_idx,:]
            loser_winner = self.crossover(loser_winner)
            loser_winner = self.mutate(loser_winner)
            self.pop[sub_pop_idx] = loser_winner

        DNA_prod = self.translateDNA(self.pop)
        fitness_now = self.get_fitness(DNA_prod)
        king = np.argmin(fitness_now)
        king_value = np.min(fitness_now)
        return king, king_value,fitness_now,DNA_prod
```
